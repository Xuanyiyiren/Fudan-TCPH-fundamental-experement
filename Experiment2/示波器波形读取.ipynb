```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
my_dict = {'s': 1, 'ms': 1e3, 'μs': 1e6 ,'ns':1e9}
# print(my_dict)

plt.rcParams.update({'axes.labelsize': 16,'xtick.labelsize': 14, 'ytick.labelsize': 14,'legend.fontsize': 12})

def mydraw(filename,title='Waveforms on the oscilloscope',label1='CH1',label2='CH2',x_label='Time',y_label='Voltage/V',yrange=((),()),time_unit='s',save=''):
    if time_unit=='us':
        time_unit='μs'
    df = pd.read_csv(filename)
    # 提取X轴和Y轴数据
    x_data = df.iloc[1:, 0].to_numpy(dtype=float)
    ch1_data = df.iloc[1:, 1].to_numpy(dtype=float)
    ch2_data = df.iloc[1:, 2].to_numpy(dtype=float)

    dt=df['Increment'][0]
    delt=df['Start'][0]
    # print('delt',delt)
    x_data=x_data*dt+delt
    x_data*=my_dict[time_unit]

    # 创建两个子图
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
    ylim1,ylim2=yrange
    # 在第一个子图中绘制第一条曲线
    ax1.plot(x_data, ch1_data, label=label1, color='orange')
    # ax1.set_xlabel(x_label)
    ax1.set_ylabel(y_label)
    if len(ylim1)!=0:
        ymin,ymax=ylim1
        ax1.set_ylim(ymin, ymax)  # 设置y轴范围为0到5
        ax1.set_yticks(np.linspace(ymin, ymax, 5))  # 设置Y轴刻度值
    ax1.legend()

    # 在第二个子图中绘制第二条曲线
    ax2.plot(x_data, ch2_data, label=label2, color='blue')
    ax2.set_xlabel(x_label)
    ax2.set_ylabel(y_label)
    if len(ylim2)!=0:
        ymin,ymax=ylim2
        ax2.set_ylim(ymin, ymax)  # 设置y轴范围为0到5
        ax2.set_yticks(np.linspace(ymin, ymax, 5))  # 设置Y轴刻度值
    ax2.legend()
    plt.xlabel(x_label+'/'+time_unit)


    # 设置标题
    plt.suptitle(title)#,fontsize=10)

    # 调整子图之间的垂直间距
    # plt.tight_layout()
    plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, hspace=0.08)
    if save!='':
        plt.savefig(save,bbox_inches='tight')

    # 显示图表
    plt.show()

    return x_data,ch1_data,ch2_data,time_unit
```

```python
import os

# 获取当前目录下的所有文件
file_list = [f for f in os.listdir() if os.path.isfile(f)]

# 按文件的最后修改时间进行排序，从上到下依次是从早到晚
sorted_file_list = sorted(file_list, key=lambda x: os.path.getmtime(x))

# 打印所有文件名
for file_name in sorted_file_list:
    if file_name[-3:]=='csv':
        print(file_name)

```

# 延迟时间的测量

```python
#序号3
y1range=y2range=(-1.5,6.5)
# save='../pic/xh3sbq.png'
t_data,ch1_data,ch2_data,timeunit=mydraw('xh3.csv',title='Measured Waveforms at the Beginning and End of a Coaxial Cable',label1='Beginning',label2='End',time_unit='ns',yrange=(y1range,y2range))
# y1range=y2range=(-6.5,6.5)
t_data,ch1_data
dif_data=ch1_data[1:]-ch1_data[:-1]
t_data=t_data[:-1]
# import matplotlib.pyplot as plt

# # 示例数据
# x = [1, 2, 3, 4, 5]
# y = [10, 8, 15, 7, 12]

# 创建折线图
plt.plot(t_data, dif_data, marker='o', linestyle='-')
plt.xlabel('Time (μs)')  # 在X轴标签中使用μ字符
plt.ylabel('Voltage')
plt.title('Line Plot with Microseconds (μs)')
plt.grid(True)
plt.show()

t_data,dif_data

# 创建一个 Pandas DataFrame，将两列数据合并为一个 DataFrame
df = pd.DataFrame({'Time/'+timeunit: t_data, 'dif_Voltage': dif_data})

# 将数据写入 Excel 文件，指定不同的列
df.to_excel("output.xlsx", index=False)
```
